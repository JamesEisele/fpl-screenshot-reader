```python
import cv2
import pytesseract
import pathlib
import os
import platform
from matplotlib import pyplot as plt
```

```python
address = "../examples/FFIFx3YXIAEL7xd.jpeg"
img = cv2.imread(address)
```

```python
def show_image(target):
    plt.figure(figsize=(15,15))
    plt.imshow(cv2.cvtColor(target, cv2.COLOR_BGR2RGB), interpolation='nearest')
    plt.show()
```

```python
show_image(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
show_image(gray)
```

```python
blurred = cv2.GaussianBlur(gray, (3,3), 0)
show_image(blurred)
```

```python
(T,thresInv) = cv2.threshold(blurred, 220, 255, cv2.THRESH_BINARY_INV)
show_image(thresInv)
```

```python
shapeMask = cv2.inRange(img, lowerb=(40,70,0), upperb=(110,160,100))
show_image(shapeMask)

inverse = cv2.bitwise_not(shapeMask)
show_image(inverse)
```

```python
combined = cv2.bitwise_and(img, cv2.cvtColor(shapeMask, cv2.COLOR_GRAY2BGR))
show_image(combined)
```

```python
print(pytesseract.image_to_string(cv2.cvtColor(combined, cv2.COLOR_BGR2RGB)))
```
